### Pauli X Gate in ground state |Î¦>=(1,0)

```python
q = QuantumRegister(1)
c = ClassicalRegister(1)
qc = QuantumCircuit(q,c)
```

```python
psi1 = Statevector(qc)
psi1.draw('latex')
```

```python
plot_state_qsphere(qc)
```

```python
rho1 = DensityMatrix(qc)
rho1.draw('latex',prefix='\\rho_1 = ')
```

```python
plot_state_city(qc)
```

```python
plot_state_hinton(qc)
```

```python
plot_state_paulivec(qc, title = 'Density matrix of the final state as linear combination of Pauli Gates') #Density_matrix=1/2*I+1/2*Pauli_z
```

```python
plot_bloch_multivector(qc)
```

```python
purity = rho1.purity()
print(purity)       #purity=Trace(Density_matrix*Density_matrix)
```

```python
qc.x(q[0])
```

```python
qc.draw()
```

```python
psi2 = Statevector(qc)
psi2.draw('latex')  # flips the ground state 
```

```python
plot_state_qsphere(qc)
```

```python
rhox = DensityMatrix(qc)
rhox.draw('latex',prefix='\\rho_x=')
```

```python
plot_state_paulivec(qc, title = 'Density matrix of the final state as linear combination of Pauli Gates') #Density_matrix=1/2*I-1/2*Pauli_z
```

```python
plot_bloch_multivector(qc)
```

```python
purityf = rhox.purity()
print(purityf)
```

```python
qc.measure(q[0],c[0])
```

```python
qc.draw()
```
